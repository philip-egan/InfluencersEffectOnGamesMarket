```python
page = requests.get("https://steamcommunity.com/market/")
page
```

```python
soup =BeautifulSoup(page.content,"html.parser")
```

```python
games=[]
for a in soup.find_all("a",{"class":"game_button"},href=True):
    games.append(a["href"])
```

```python
print(a)
```

```python
appids=[]
for x in games:
    appids.append(re.findall(r'\d+',x))
```

```python
print(len(appids))
```

```python
with open('marketplaceFiles/ccudata/.txt', 'w') as filehandle:
    for aID in appids:
        filehandle.write('%s\n' % aID)
```
