creates csv files from the price history for easier usage

```python
onlyfiles = [f for f in listdir("marketplaceFiles/priceHistory/1567030/") if isfile(join("marketplaceFiles/priceHistory/1567030/", f))]
print(len(onlyfiles))
```

```python
my_file = open("marketplaceFiles/appIds-Copy1.txt", "r")
content = my_file.read()
content_list = content.split("\n")
my_file.close()
```

```python
print(content_list[2])
```

```python
'''
this takes the raw market figures them and cleans them into a usable form removing unwanted parts of the string
startfile is start location loop
endfile is the loop ending
fileN the game that will be looked at
'''
def csvtoFile(startfile,endfile,fileN):
    for filecount in range(startfile,endfile,48):
        if os.stat("marketplaceFiles/priceHistory/"+str(content_list[fileN])+"/"+onlyfiles[filecount]).st_size==0:
            print(onlyfiles[filecount])
        with open("marketplaceFiles/priceHistory/"+str(content_list[fileN])+"/"+onlyfiles[filecount], encoding='utf8') as f:
                content=f.readlines()
        datalist=[]
        for x in content:
            a=str(x)
            b1=a.replace("[","")
            b2=b1.replace("]","")
            b3=b2.replace("\n","")
            b3=b3.replace("'","")
            datalist.append(b3)
        df=pd.DataFrame.from_records([sub.split(",")for sub in datalist],columns=["date","price","volume"])
        for datecount in range(0,len(df["date"])):
            df["date"][datecount]=datetime.strptime(df["date"][datecount],"%b %d %Y %H: +0")
            df["date"][datecount]=datetime.strftime(df["date"][datecount],"%b %d %Y")
        df[str(onlyfiles[filecount])]=df["price"].astype(float)*df["volume"].astype(float)
        df.to_csv("marketplaceFiles/marketcsv/"+str(content_list[fileN])+"/"+onlyfiles[filecount],index=False)
        print(filecount)
```
