move all games that have player infomation, this is done to remove games that had no player information of which there were many.
the data was put into a csv format

```python
import csv
import json
import threading
import pandas as pd
from os import listdir
from os.path import isfile, join
from glob import glob
import os
import shutil
```

```python
'''
this gives a list of all files in the foulder to and outs them into the series onlyfiles
'''
onlyfiles = [f for f in listdir("marketplaceFiles/ccupeakdata/") if isfile(join("marketplaceFiles/ccupeakdata/", f))]
print(len(onlyfiles))
```

```python
'''
this takes all files from ccupeakdata and checks the files sizes
if the file size is 2 bits the file is ignored
if it is more than 2 bits it is replicated in actual player data
'''
for filecount in range(0,len(onlyfiles)):
    if os.stat("marketplaceFiles/ccupeakdata/"+onlyfiles[filecount]).st_size==2:
    else:          
        shutil.copy("marketplaceFiles/ccupeakdata/"+str(onlyfiles[filecount]),"marketplaceFiles/actualPlayerData/"+str(onlyfiles[filecount]))
```

```python
'''
this gives a list of all files in the foulder to and outs them into the series onlyfiles2
'''
onlyfiles2 = [f for f in listdir("marketplaceFiles/actualPlayerData/") if isfile(join("marketplaceFiles/actualPlayerData/", f))]
print(len(onlyfiles2))
```

turning this data into a csv

```python
for filecount2 in range(0,len(onlyfiles2)):
    with open('marketplaceFiles/actualPlayerData/'+str(onlyfiles2[filecount2])) as json_file:
        data = json.load(json_file) 
    df=pd.DataFrame.from_dict(data)
    df=df.apply(pd.Series)
    df=df.transpose()
    df.columns=["Peak Players","Average Players"]
    df.index=pd.to_datetime(df.index, format="%B %Y")
    df.to_csv("marketplaceFiles/actualPlayerData/csvActualPlayerData/"+str(onlyfiles2[filecount2]))
```
